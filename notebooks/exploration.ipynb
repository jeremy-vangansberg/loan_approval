```python
X = df.copy()
```

```python
X = X.drop(columns=['Name', 'LoanNr_ChkDgt', 'Bank', 
                     'BankState', 'DisbursementDate', 'ApprovalDate', 
                    'ApprovalFY', 'FranchiseCode', 'Zip', 'City', 'State', 'ChgOffDate', 'SBA_Appv'])
```

```python
X=X.dropna()
```

```python
columns = ['DisbursementGross','BalanceGross','ChgOffPrinGr','GrAppv']
for column in columns: 
    X[column] = X[column].str.replace('$', '').str.replace(',','').str.strip()
```

```python
X.info()
```

```python
X[columns] = X[columns].astype('float64').astype('int64')
```

```python
values = ['0','T','1','R','`','2','C','3',',','7','A','5','.','4','-','Q']
X = X[X.RevLineCr.isin(values) == False]
```

```python
values = ['0','C','S','A','R','1']
X = X[X.LowDoc.isin(values) == False]
```

```python
cat = ['RevLineCr', 'LowDoc']
```

```python
X = X.drop('NAICS', axis =1)
```

```python
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures, StandardScaler, RobustScaler
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import numpy as np
from scipy.stats import kruskal
from sklearn.model_selection import cross_validate
```

```python
X.columns
```

```python
y = X['MIS_Status'].astype('category').cat.codes
X = X.drop(['MIS_Status'], axis=1)

num = list(X.select_dtypes(include=['int', 'float']).columns)
```

```python
num
```

```python
std = StandardScaler()
ohe = OneHotEncoder()

preprocessor = make_column_transformer(
    (ohe, cat), remainder='passthrough'
    # (std, num)
)
```

```python
cat
```

```python
X.columns
```

```python
X.info()
```

```python
X.loc[ X.NewExist == 0, 'NewExist'] = np.nan
```

```python
X['NewExist']  = X.NewExist.replace({2. : True, 1.0 : False})
```

```python
X.NewExist.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.1, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(class_weight='balanced', max_depth=5)
rfc_pipe = make_pipeline(preprocessor, rfc)

rfc_pipe.fit(X_train, y_train)
```

```python
rfc_pipe.score(X_test, y_test)
```

```python
import seaborn as sns
```

```python
x_to_plot = rfc_pipe[-1].feature_importances_
y_to_plot = rfc_pipe[:-1].get_feature_names_out()

sns.barplot(x=x_to_plot,y=y_to_plot)
```
